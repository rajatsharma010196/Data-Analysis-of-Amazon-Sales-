# Data Cleaning and Manipulation Steps (Reading Assignment)

We have done all the data cleaning and manipulation steps below:

Dropped 1 column:


```python
# drop column Order Id

sales_records.drop(['Order ID'],axis = 1, inplace = True )
```

```python
sales_records.head()
```

We will now deal with time related columns, we have two time related columns

Order Date

Ship Date

We will first convert these column to datatime data type of pandas.

```python
sales_records['Order Date'] = pd.to_datetime(sales_records['Order Date'])
sales_records['Ship Date'] = pd.to_datetime(sales_records['Ship Date'])
```

```python
sales_records.head()
```

```python
sales_records.info()

# changes reflecting in info
```

We will create new features from these

1. ship day name - this is basically the day of the week when order took place

2. Shipping duration - Total days spent for shipping

```python
# create 'Shipping duration' column using np.timedelta64(1,'D')
sales_records['Shipping duration'] = (sales_records['Ship Date'] - sales_records['Order Date'])/np.timedelta64(1,'D')

# create 'Ship day name' column using 'Ship Date' column using day_name()
sales_records['Ship day name'] = sales_records['Ship Date'].dt.day_name()

# print data info
print(sales_records.info())

# print data head
sales_records.head()
```

Let's also see the number of missing values for each column

```python
# as the profit depends on the difference of unit selling price and unit cost price, we can create a new column
# for this difference as Unit Profit

sales_records['Unit Profit'] = round(sales_records['Unit Price'] - sales_records['Unit Cost'],2)

# using round command to round off it to two decimal places
```

```python
sales_records.head()
```

```python
# print missing values for each column - use .isnull().sum
sales_records.isnull().sum(axis=0).reset_index()
```

From the table we observe that we have no missing values

We have 'Order Priority' column given in terms of letters H, C, L, M

Assuming H - High, C - Critical, L - Low, M- Medium priorities

We will replace the column with these strings

```python
# defining a function for this

def order_priority(x):
    if x=='H':
        return 'High'
    elif x=='C':
        return 'Critical'
    elif x=='L':
        return 'Low'
    elif x=='M':
        return 'Medium'
    
# use .apply to the order priority column to apply the function created
    
sales_records['Order Priority'] = sales_records['Order Priority'].apply(order_priority)
```

Since we are trying to understand the distribution of continuous numerical variables, we will be using 
* histograms
* box plots

Below we have used a for loop to loop through all the continuous variables and then draw histograms and box plots for each of them at each iteration.

```python
# for loop for continuous_columns variable
for feature in continuous_columns:
    # Ignore warnings
    import warnings
    warnings.filterwarnings('ignore')
    # We created two variables, fig and ax
    # These two variables now hold the two core objects used for all types of plotting operations. 
    fig,ax = plt.subplots(1,2,figsize=(12,6))
    #     First object fig, short for figure, imagine it as the frame of your plot. 
#     You can resize, reshape the frame but you cannot draw on it. 
#     On a single notebook or a script, you can have multiple figures. 
#     Each figure can have multiple subplots. Here, subplot is synonymous with axes. 
#     The second object, ax, short for axes, is the canvas you draw on. 
#     Or rephrasing, it is the blank sheet you can plot and hold your data. 
#     An axes object can only belong to one figure.
    ax[0].hist(sales_records[feature])
    ax[0].set_title('histogram of column values in '+feature)
    sns.boxplot(sales_records[feature],ax=ax[1])
    # using ax2.set_title for box plot
    ax[1].set_title('box plot of column values in '+feature)
    # seaborn style setting
    sns.set(style='ticks', palette='deep', font_scale=1.1)
    # matplotlib command for displaying plots
    plt.show()
```

From the box plot we can see we have more than 3 outliers for this column with maximum value of 1.7e+06 which is a number more than the average total revenue.

### Total Profit vs Month

```python
# applying groupby() function to group the data into month column

df_month = sales_records.groupby(['Month'])


# now we need to find the sum of other feilds to make the comparison

df_month = df_month.sum().reset_index()

df_month
```

```python
# let us sort the month name

df_month = df_month.sort_values('Month', key = lambda x : pd.to_datetime(x, format='%b').dt.month)

df_month
```

```python
# now the data is ready we can make the required plots


fig, ax = plt.subplots(figsize=(15,7))

sns.lineplot(x = df_month['Month'],y = df_month['Total Profit'], data = df_month, ax = ax)

ax.set_title('Line Graph for Total Profit vs Month')

plt.show()
```

From the plot peak we can see maximum total profit appears in months Feb and Nov , and least in Mar and Aug
